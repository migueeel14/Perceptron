# Introduccion
El cerebro humano realiza tareas sofisticadas de procesamiento de información, utilizando hardware y reglas de operación muy diferentes de las que se basan en las computadoras convencionales. Los procesadores en el cerebro son las neuronas, elementos bastante ruidosos que operan en paralelo. Están organizados en redes densas, cuya estructura puede variar de muy regular a casi amorfo (ver figura 2), y comunican señales a través de una gran cantidad de conexiones entre ellas (las llamadas sinapsis). Estas conexiones representar el "programa" de una red. Al actualizar continuamente las fortalezas de las conexiones, una red en su conjunto puede modificar y optimizar su "programa", "aprender" de la experiencia y adaptarse a las circunstancias cambiantes.

Desde el punto de vista de la ingeniería, las neuronas son procesadores bastante pobres, son lentos y poco confiables (ver la tabla a continuación). Sin embargo, en el cerebro siempre hay una gran cantidad de neuronas involucradas en cualquier tarea operando en paralelo con muchas conexiones entre ellas. Lo que escompletamente opuesto a las computadoras convencionales, las cuales, generalmente, realizan las operaciones en paralelo. Esto implica que la falla de cualquier parte de la cadena de operaciones es usualmente fatal. Adicionalemnte, las computadoras convencionales ejecutan una especificación detallada de
órdenes, que requieren que el programador sepa exactamente qué datos se pueden esperar y como responder. Cambios no previstos por el programador pueden conducir a problemas. Las redes neuronales, por otro lado, puede adaptarse a circunstancias cambiantes.

Es necesario entender los principios detrás del procesamiento de la información neuronal para poder diseñar software adaptativo y sistemas de procesamiento de información artificial que pueda 'aprender'. Para esto se utilizan modelos de redes de neuronas muy simplificados. Como sus contrapartes biológicas, estos sistemas artificiales, las conexiones entre neuronas son 'entrenadas' en lugar de ser programadas. Las neuronas "aprenden" a realizar tareas al presentarles ejemplos de lo que se supone que deben hacer. Es necesario entender las relaciones entre el rendimiento de la red para un tipo dado de tarea, la
elección de 'regla de aprendizaje' (la receta para la modificación de las conexiones) y la arquitectura de red. 
Adicionalmente, es conveniente explotar el eficiente procesamiento de informacion que tienen las redes neuronales reales (biológicas), mediante la construcción de redes neuronales artificiales que también funcionan en paralelo. Estos sistemas, tienen el potencial de ser máquinas de procesamiento de información increíblemente rápidas.
Finalmente, debido a su estructura compleja, la gran cantidad de elementos involucrados y su naturaleza dinámica, los modelos de redes neuronales exhiben un comportamiento altamente no trivial.


Las neuronas operan más o menos de la siguiente manera: La membrana celular de una neurona mantiene diferencias de concentración de varios iones (los principales son Na+, K+ y Cl) entre el interior y el exterior de la célula, por medio de acciones combinadas de bombas de iones activos y canales de iones controlables. Cuando la neurona está en reposo, los canales están cerrados y debido a la actividad de las bombas y las diferencias de concentración, el interior de la neurona tiene un potencial eléctrico negativo de alrededor de 70 mV, en comparación con el liquido exterior. Sin embargo, excitación eléctrica local suficientemente fuerte que haga el potencial celular
temporalmente menos negativo, conduce a la apertura de canales específicos, lo que a su vez provoca una reacción en cadena de otros canales que se abren y/o cierran, con una resultado neto la generación de un pico eléctrico de altura alrededor de +40 mV, con una duración de aproximadamente 1 ms, que se propagará a lo largo de la membrana a una velocidad de aproximadamente 5 m/seg: el llamado potencial de acción. Después de esta avalancha electroquímica se necesitan unos pocos milisegundos para restaurar la paz y el orden. Durante esto
período, el llamado período de rstauracion, la membrana solo puede generar un potencial de acción por una excitación extremadamente fuerte. El potencial de acción sirve como señal de comunicación eléctrica, propagándose y bifurcandose a lo largo del canal de salida de la neurona, el axón, hacia las otras neuronas. Ya que la propagación de un potencial de acción a lo largo de un axón es el resultado de un proceso electro/químico activo,  la señal retendrá forma y fuerza, incluso después de la bifurcación. La unión entre un canal de salida (axón) de una neurona y una entrada canal (dendrita) de otra neurona, se llama sinapsis.


La representacion más simple de una neurona es aquella en la que sus posibles estados se reducen a una sola variable binaria S, que indica si esta activa (S = 1) o está en reposo (S = 0). 
Si la senal total recibida por la neurona es mayor que el potencial de activacion de la misma $\theta$, determina si la neurona esta activa o no. Se asume que las senales de entrada se suman linealmente, ponderadas por valores asociados a las sinapsis, representadas por $w_i$.  

Las redes neuronales artificiales fueron concebidas originalmente como un modelo que imitaría la función de las neuronas biologicas.
En este contexto, son una suma ponderada de entradas, en algunos casos, seguida de una función de activación lineal o no lineal aplicada a una neurona. 
La neurona McCulloch-Pitts de 1943 es uno de los primeros ejemplos de una neurona artificial, fuertemente influenciada por los patrones de activacion de las neuronas biologicas. Algunas redes neuranales de una capa como el perceptrón de 1958, actualizan los pesos $w_i$ con una regla de aprendizaje. 

Los elementos basicos de una red neuronal son los siguentes:
* **Entradas o caracteristicas**: 
* **Pesos $w_i$** , con $i={1:m}$: son numeros reales que determinan la importancia de cada entrada en la salida final. Cuanto mayor es el peso, mayor es la importancia de la variable de entrada. 
* **Bias o sesgo $b$**: El bias también es un peso, esta relacionado con los factores que no se han considerado, factores imprevistos o no observables en los datos de entrada. 
* **Salidas**: dado un conjunto de $n$ datos con características $x_i$ con $i={1:m}$, y un conjunto de pesos $w_i$ con $i={1:m}$, se produce una salida binaria con valores $y=1$  ó $y=0$. Esta salida depende de la suma de la multiplicacion de las características de los datos de entrada y los pesos $z = \Sigma_i m w_i*x_i$, así como del sesgo $b$.
* **Funcion de activacion** $f(z)$: Despues de sumar todas las entradas con sus respectivos pesos, se obtiene una variable $z$, a partir de esta variable se debe tomar una decisión que dará la variable de salida. Esto se hace a través de una función de activación representada como $f (z)$. Esta función puede ser lineal y no lineal.
*La elección de la función depende en gran medida del problema a resolver o de lo que se espera que la NN aprenda.*

* Funciones de activación lineal 
\begin{equation}
    f(z) = 
    \begin{cases}
      0 & \text{if   }\quad  z\leq 0\\
      1 & \text{if   }\quad  z \gt 0
    \end{cases}
\end{equation}
$$ $$ $$f(z) = sgn(z) =\begin{cases}
    1 & \text{if }z \gt 0\\
    -1 & \text{if }z \lt 0.
  \end{cases} $$ $$ $$
  
$$f (z) = z. $$


* Unidades lineales rectificadas (Rectified Linear Units, ReLU): con ReLU, la salida no es menor a cero (o negativa). Por lo tanto, si z es mayor que cero, nuestra salida sigue siendo z; de lo contrario, si z es negativa, la salida es cero. La fórmula es
$$f (z) = max (0, z).$$ 
* Tanh:  la función de activación es la tangente hiperbólica de z, dada por
$$f (z) = tanh (z)$$
* Sigmoide: esta  función de activación esta dada por la siguiente fórmula:
$$f (z) =\frac{1}{1 + e ^ {-  z}}.$$

* **Función de salida:** el valor resultante de esta función es la salida de las neuronas, esta función determina que valor se transfiere a las neuronas vinculadas. Los valores de salida pueden ser binarios {0, 1} o {-1, 1}, o están comprendidos en el rango [0, 1] o [-1, 1].
Las funciones de salida mas comunes son:
* Ninguna: este es el tipo de función más sencillo, la salida es la misma que la entrada, también es llamada función identidad.
* Binaria: depende de un umbral determinado $\theta$
$$ \phi(z) =\begin{cases}
    1 & \text{if }f(z) \gt \theta\\
    0 & \text{otherwise}.
  \end{cases} $$  $$ $$

**Notación**
- $N$ tamaño de la muestra o características

- $m_{train}$ número de muestras de entrenamiento
- $m_{test}$ número de muestras de entrenamiento
- $\mathbf{x} \in \mathbb{R}^N,$ es el vector de características (features vector)
\begin{equation} 
X = \begin{bmatrix}

\end{bmatrix} _{N \times m}, \quad \mathbf{x} \in \mathbb{R}^N.
\end{equation} 
- $\mathbf{w} \in \mathbb{R}^N,$ es el vector de pesos
\begin{equation} 
W = \begin{bmatrix} 
 & &\\ \mathbf{w}^{(1)} & \mathbf{w}^{(2)} & \cdots & \mathbf{w}^{(m)}\\ & &
\end{bmatrix} _{N \times m}, \quad \mathbf{w} \in \mathbb{R}^N.
\end{equation}
